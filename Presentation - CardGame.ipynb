# Card Game

## My Card Game includes 3 cards per person. You can choose how many rounds (at least 1 round) and players (at least 2 people) you want.

```python
import random
from tabulate import tabulate
```

```python
class Card:    # to combine rank with suit to create 1 card
    suit_list =["Clubs", "Diamond", "Hearts", "Spades"]
    rank_list = ["None", "Ace", "2", "3", "4", "5", "6","7","8","9","10","Jack","Queen","King"]
    def __init__(self, suit = 0, rank = 1):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return (self.rank_list[self.rank] + " of " + self.suit_list[self.suit])
    
print(Card(3,5))
```

```python
class Deck:    # to create 52 cards of a deck after that shuffle and deal cards for players
    def __init__(self):
        self.cards = []
 
        for suit in range(4):
            for rank in range(1,14):
                self.cards.append(Card(suit,rank))               
                
    def __str__(self):
        s = ""
        for i in range(len(self.cards)):
            s += i * " " + str(self.cards[i]) + "\n"
        return s
    
    def shuffle(self):          # to shuffle cards
        n_cards = len(self.cards)
        for i in range(n_cards):
            j = random.randrange(0, n_cards)
            self.cards[i], self.cards[j] = self.cards[j], self.cards[i]
            
    def pop_card(self):        # to remove a card
        return self.cards.pop()
    
    
    def deal(self, hands, n_cards= 52):    # to deal cards for players
        n_players = len(hands)
        for i in range(n_cards):
            card = self.pop_card()
            current_player = i % n_players
            hands[current_player].add_card(card)
```

```python
class Hand(Deck):     # to combine the player's name with 3 cards
    def __init__(self, name = ""):
        self.cards = []
        self.name = name
        
    def add_card(self,card):
        self.cards.append(card)
        
    def __str__(self):
        s = "**  Hand of " + self.name + " contains \n" + Deck.__str__(self)
        return s
```

```python

class CardGame(Hand):   # This is my card game
    def __init__(self):
        self.deck = Deck()
        self.deck.shuffle()
        
    def rule_game(self):
        
        list_cards = []     # to take the rank of cards
        for i in range(len(self.hands)):
            list_cards.append([])
            for t in range(3):
                list_cards[i].append(self.hands[i].cards[t].rank)
        # print(list_cards)
        
        
    # Rule 1: There may have many people win. (Consider Jack, Queen and King cards)
   
        rule_1 = []
        rule_2 = []
        total=[]
        for k in range(len(self.hands)):
            # to find the players have 3 cards form Jack to King
            if list_cards[k][0] >= 11 and list_cards[k][1] >= 11 and list_cards[k][2] >= 11:
                rule_1.append(k)
                print( "----> Player " + self.hands[k].name + " WINS.\n")
    
    # Rule 2 (if rule 1 have no result):There may have many people win. (Consider 2 numbers of the sum of 3 cards)
        
        if not rule_1:
            list_totals = []
            list_unit_numbers = []
            for j in range(len(self.hands)):
                for m in range(3):
                    # to change the number of Jack, Queen and King into 10 points
                    if list_cards[j][m] > 10:
                        list_cards[j][m] = 10
                    # to split the first number and the unit/second number after sum
                list_totals.append(list(reversed(str(sum(list_cards[j])))))  
                # to take the unit number
                list_unit_numbers.append(list_totals[j][0])

         
            for l in range(len(self.hands)):
                if list_unit_numbers[l]== max(list_unit_numbers):
                    rule_2.append(l)  # to find how many players have the same point with the largest point
            if len(rule_2) == 1:
                print("----> Player " + self.hands[rule_2[0]].name + " WINS.\n")
                
            # if exceed one person win, we continue to compare the first number of a number after sum
            
            elif len(rule_2)>1:       # there is a error when submit file
                list_first_numbers = []    # to find the first number of a number after sum
                list_names = []            # to find the player's name 
                for c in rule_2:
                    if len(list_totals[c])==1:          # to add the first number 0 if the sum has 1 number
                        list_first_numbers.append(0)
                    else:
                        list_first_numbers.append(int(list_totals[c][1]))
                    list_names.append(c)
                D = dict(zip(list_names,list_first_numbers))
                  # to find how many winners after alls
                for name in D.keys():
                    if D[name] == D[max(D,key=D.get)]:
                        total.append(name)
                        print("----> Player " + self.hands[int(name)].name + " WINS.\n")
                        

                
            # print(list_totals)  
                
        return [rule_1,rule_2,total]
```
